# AI-Based Market Trend Analysis and Demand Forecasting
Dataset: Google Trends (Walmart – US)
Model: Random Forest Regressor


## 1. Problem Definition

Accurate demand forecasting is critical for retail businesses to support inventory management, pricing, and supply-chain planning.  
This project aims to analyze market trends and predict future demand using Google Trends data and machine learning techniques.


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv(
    "../data/google_trends.csv",
    skiprows=1
)

df.columns = ["date", "demand"]

df["date"] = pd.to_datetime(df["date"])
df["demand"] = pd.to_numeric(df["demand"], errors="coerce")

df = df.dropna()

df.head()
```

Google Trends exports contain metadata rows.
We skipped non-data rows and standardized the dataset into
a clean time series with date and demand columns.


## 2. Data Understanding & Preparation

Google Trends provides weekly normalized search interest values on a scale of 0–100, representing relative consumer demand.  
The raw dataset contains metadata rows, which were removed to obtain a clean time-series suitable for analysis and forecasting.


```python
df["t"] = range(len(df))
df["lag_1"] = df["demand"].shift(1)
df["lag_2"] = df["demand"].shift(2)
df["lag_4"] = df["demand"].shift(4)
df["rolling_4"] = df["demand"].rolling(4).mean()

df = df.dropna()
df.head()
```

Lag and rolling features were created to capture short-term momentum and seasonal demand patterns.
These features help the model learn temporal dependencies in market demand.


## 3. Feature Engineering

Temporal features were engineered to capture demand dynamics, including lag variables and rolling averages.  
These features enable the model to learn short-term momentum and medium-term seasonal patterns in market demand.


```python
FEATURES = ["t", "lag_1", "lag_2", "lag_4", "rolling_4"]

split = int(len(df) * 0.8)
train = df.iloc[:split]
test = df.iloc[split:]

X_train = train[FEATURES]
y_train = train["demand"]

X_test = test[FEATURES]
y_test = test["demand"]
```

## 4. Model Design & Justification

A Random Forest Regressor was selected for demand forecasting due to its ability to model non-linear relationships and handle noisy real-world data.  
Compared to linear models, Random Forest provides greater flexibility, while avoiding the complexity and data requirements of deep learning models.


```python
model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X_train, y_train)
```

## 5. Training & Evaluation

The dataset was split using a time-based train–test strategy to simulate real-world forecasting conditions.  
Model performance was evaluated using Mean Absolute Error (MAE), which measures the average deviation between predicted and actual demand values.


```python
preds = model.predict(X_test)
mae = mean_absolute_error(y_test, preds)
mae
```

## 6. Evaluation & Analysis

The trained model achieved an MAE of approximately 3.23, indicating strong predictive accuracy on unseen data.  
Given the normalized demand scale, this error level demonstrates that the model effectively captures underlying demand trends while smoothing short-term volatility.


```python
plt.figure(figsize=(12,5))
plt.plot(test["date"], y_test.values, label="Actual")
plt.plot(test["date"], preds, label="Predicted")
plt.title(f"Demand Prediction (MAE = {mae:.2f})")
plt.legend()
plt.show()
```

## 7. Future Demand Forecast

The trained model was used to generate demand forecasts for the next 12 weeks using a recursive prediction strategy.  
This approach allows future demand estimates to be generated based solely on historical patterns learned by the model.


```python
future = df.tail(12).copy()
future_preds = model.predict(
    future[["t",  "lag_1", "lag_2", "lag_4", "rolling_4"]]
)
future_preds
```

```python
n_weeks = 12

last = df.iloc[-1]

current_t = last["t"]
lag_1 = last["demand"]
lag_2 = df.iloc[-2]["demand"]
lag_4 = df.iloc[-4]["demand"]
rolling_4 = df.iloc[-4:]["demand"].mean()

future_preds = []

for _ in range(n_weeks):
    current_t += 1

    X_future = pd.DataFrame([{
        "t": current_t,
        "lag_1": lag_1,
        "lag_2": lag_2,
        "lag_4": lag_4,
        "rolling_4": rolling_4
    }])

    pred = model.predict(X_future)[0]
    future_preds.append(pred)

    # shift lags forward
    lag_4 = lag_2
    lag_2 = lag_1
    lag_1 = pred
    rolling_4 = (rolling_4 * 3 + pred) / 4

future_preds

import pandas as pd

# Number of weeks predicted
n_weeks = len(future_preds)

# Start from next week after last known date
start_date = df["date"].iloc[-1] + pd.Timedelta(weeks=1)

forecast_df = pd.DataFrame({
    "Week": pd.date_range(start=start_date, periods=n_weeks, freq="W"),
    "Predicted_Demand": future_preds
})

forecast_df

```

An 8-week moving average was used to smooth weekly demand fluctuations and highlight the underlying market trend. This improves interpretability without affecting the model training process.


```python
import matplotlib.pyplot as plt
import pandas as pd

# Dates for forecast
future_dates = pd.date_range(
    start=df["date"].iloc[-1] + pd.Timedelta(weeks=1),
    periods=len(future_preds),
    freq="W"
)

# Rolling mean for smoother trend
df["rolling_trend"] = df["demand"].rolling(8).mean()

plt.figure(figsize=(14,6))

# Historical demand
plt.plot(
    df["date"], df["demand"],
    label="Historical Demand",
    color="steelblue",
    alpha=0.6
)

# Smoothed trend
plt.plot(
    df["date"], df["rolling_trend"],
    label="Historical Trend (8-week MA)",
    color="navy",
    linewidth=2
)

# Test predictions (optional but strong)
plt.plot(
    test["date"], preds,
    label="Model Predictions (Test Period)",
    color="orange",
    linestyle="--",
    linewidth=2
)


# Forecast
plt.plot(
    future_dates, future_preds,
    label="12-Week Forecast",
    color="green",
    linestyle="--",
    linewidth=2
)

# Vertical line = forecast start
plt.axvline(
    x=df["date"].iloc[-1],
    color="gray",
    linestyle=":",
    label="Forecast Start"
)

plt.title("AI-Based Market Demand Forecast (Walmart)", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Demand Index (Google Trends)")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

The predictions produced by this system are intended to support decision-making and should not be used as definitive outcomes.
